```
%matplotlib inline
import numpy as np
x = [10,10,0]
j = 0
for i in x:
  if j == 0:
    x1 = i
  if j == 1:
    x2 = i
  if j == 2:
    x3 = i
  
  j = j + 1

c = np.sqrt(x1**2+x2**2+x3**2)
print (c)
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-15,15,150)
y = np.linspace(0,5,50)
y = 3*x + 1
y2 = (-1/3)*x +1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```
%matplotlib inline
import math
import matplotlib.pyplot as plt
x = []
y = []
y2 = []
for i in range (-10000,10000):
  x_ = i/10000
  x.append(x_)
  y.append(math.sqrt(1 - x_**2))
  y2.append(-math.sqrt(1 - x_**2)) 
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```
%matplotlib inline
import math
import matplotlib.pyplot as plt
x = []
y = []
y2 = []
k = 1/2
for i in range (-10000,10000):
  x_ = i/10000
  x.append(x_)
  y.append(math.sqrt(2 - (x_**2)*k))
  y2.append(-math.sqrt(2 - (x_**2)*k)) 
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
